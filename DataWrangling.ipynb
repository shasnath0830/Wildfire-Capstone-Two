### ðŸ” Handling Duplicate Wildfire Records

After reviewing the dataset for duplicate entries, intially there were no exact duplicate rows. However, further inspection revealed multiple wildfire records with identical `Location` and `Started` timestamps, which is highly unlikely in real-world scenarios.

Since two wildfires cannot occur at the exact same location and start time, we treated these as redundant records. To ensure the integrity and accuracy of the data for modeling, we decided to remove duplicates based on the combination of `Location` and `Started`, keeping only the first occurrence of each.

This step helps eliminate noise and potential bias in our binary classification model, where we aim to predict wildfire occurrences based on temporal and spatial patterns.

Since the current dataset contains only records of actual wildfire evetns `WildfireOccured` = 1, we need to simulate realistic non-wildfire scenarioos `WildfireOccured` = 0 for binary classification. In this process I'll be generating locations and dates that did not expierence wildfires the timeframe and area covered in the orignal dataset.

```python
# %pip install geopandas
```

```python
import geopandas as gpd
gdf = gpd.read_file("California_Counties.geojson")
```

```python
gdf.head()
```

```python
gdf['Counties'] = gdf['NAME'].str.replace(' County', '', regex=False)
```

```python
gdf[['Counties', 'geometry']].head()
```

```python
import random
from shapely.geometry import Point
from datetime import datetime, timedelta
import pandas as pd

# 1. Prep geometry + wildfire dates
valid_gdf      = gdf[['Counties','geometry']].copy()
wildfire_dates = set(wildfire_data['Started'].dt.date)

# 2. Helper to sample a point inside any (multi)polygon
def random_point_in_geom(geom):
    minx, miny, maxx, maxy = geom.bounds
    while True:
        x, y = random.uniform(minx, maxx), random.uniform(miny, maxy)
        p = Point(x, y)
        if geom.contains(p):
            return p

# 3. Define date window & total days
start_dt   = datetime(2013, 1, 1)
end_dt     = datetime(2019,12,31)
total_days = (end_dt - start_dt).days

# 4. Figure out which counties to skip entirely
skip_counties = {'Mexico', 'State of Oregon', 'State of Nevada'}

# 5. Count real wildfires per county (but drop the three we donâ€™t want)
wf_counts = (
    wildfire_data.loc[~wildfire_data['Counties'].isin(skip_counties), 'Counties'].value_counts()
)

# 6. Simulate non-fire events proportional to real counts
sim_rows = []
for county, n_sim in wf_counts.items():
    # pull this countyâ€™s polygon
    geom = valid_gdf.loc[valid_gdf['Counties'] == county, 'geometry']
    if geom.empty:
        # should not happen now, since skip_counties removed the outâ€ofâ€bounds ones
        continue
    geom = geom.iloc[0]

    for _ in range(n_sim):
        # a) random point inside that county
        pt = random_point_in_geom(geom)

        # b) random datetime not matching any real-fire date
        while True:
            cand = start_dt + timedelta(
                days   = random.randrange(total_days),
                hours  = random.randrange(24),
                minutes= random.randrange(60)
            )
            if cand.date() not in wildfire_dates:
                rand_dt = cand
                break

        # c) build the record
        sim_rows.append({
            "Latitude":         round(pt.y, 6),
            "Longitude":        round(pt.x, 6),
            "Started":          rand_dt.strftime("%Y-%m-%d %H:%M:%S+00:00"),
            "AcresBurned":      0,
            "Counties":         county,
            "WildfireOccurred": 0
        })

simulated_df = pd.DataFrame(sim_rows)
```
